```python
disease_names = []
```

```python
for i, link in enumerate(link_list):
    # if i == 10:
    #     break

    r = requests.get(link + "#clinical_significance")
    text = r.text
    # Parse the HTML content
    soup = BeautifulSoup(text, 'html.parser')
    # Find the table by its ID
    table = soup.find('table', id='clinical_significance_datatable')
    # Extract the rows from the table body
    rows = table.find('tbody').find_all('tr')

    # Extract the data from each row
    table_data = []
    for row in rows:
        cols = row.find_all('td')
        cols = [col.get_text(strip=True) for col in cols]
        table_data.append(cols)
    
    # Print the extracted data
    names = []
    for entry in table_data:
        print(entry[1])
        names.append(entry[1])
    
    disease_names.append([names, link + "#clinical_significance"])
```

```python
print(disease_names)
print(len(disease_names))
```

```python
# Fill cvs file with information
import csv

header = ["Disease Name", "Link"]

with open('clinical_significance.csv', 'w', newline='') as file:
    writer = csv.writer(file)

    writer.writerow(header)
    for row in disease_names:
        sanitized_names = ''
        if len(row[0]) == 1:
            for name in row[0]:
                sanitized_names = name
        else:
            for name in row[0]:
                sanitized_names += name
                sanitized_names += ", "
        row_to_write = [sanitized_names, row[1]]
        writer.writerow(row_to_write)
```
